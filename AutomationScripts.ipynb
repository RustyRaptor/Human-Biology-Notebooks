```python
from PyPDF2 import PdfReader, PdfWriter

def append_pdfs(input_pdf1, input_pdf2, output_pdf):
    # Create PdfReader objects for both input PDFs
    pdf1 = PdfReader(input_pdf1)
    pdf2 = PdfReader(input_pdf2)

    # Create a PdfFileWriter object for the output PDF
    pdf_writer = PdfWriter()

    # Append pages from the first PDF
    for page_num in range(len(pdf1.pages)):
        pdf_writer.add_page(pdf1.pages[page_num])

    # Append pages from the second PDF
    for page_num in range(len(pdf2.pages)):
        pdf_writer.add_page(pdf2.pages[page_num])

    # Write the combined PDF to the output file
    with open(output_pdf, 'wb') as output_file:
        pdf_writer.write(output_file)
        
def compile_lab_report(first_page, appendix_a, appendix_b, output_pdf):
        # Create PdfReader objects for all input PDFs
        pdf1 = PdfReader(first_page)
        pdf2 = PdfReader(appendix_a)
        pdf3 = PdfReader(appendix_b)
        
        # Create a PdfFileWriter object for the output PDF
        pdf_writer = PdfWriter()
        
        # Append pages from the first PDF
        for page_num in range(len(pdf1.pages)):
                pdf_writer.add_page(pdf1.pages[page_num])
        
        # Append pages from the second PDF
        for page_num in range(len(pdf2.pages)):
                pdf_writer.add_page(pdf2.pages[page_num])
        
        # Append pages from the third PDF
        for page_num in range(len(pdf3.pages)):
                pdf_writer.add_page(pdf3.pages[page_num])
        
        # Write the combined PDF to the output file
        with open(output_pdf, 'wb') as output_file:
                pdf_writer.write(output_file)
                
        
def compile_all_pdfs(list_of_pdfs):
        # Create a PdfFileWriter object for the output PDF
        pdf_writer = PdfWriter()
        
        # Iterate over the list of PDFs
        for pdf_file in list_of_pdfs:
                # Create a PdfReader object for the current PDF
                pdf = PdfReader(pdf_file)
                
                # Append each page from the current PDF
                for page_num in range(len(pdf.pages)):
                        pdf_writer.add_page(pdf.pages[page_num])
        
        # Write the combined PDF to the output file
        with open('compiled_pdf.pdf', 'wb') as output_file:
                pdf_writer.write(output_file)
        

# # compile lab report
# compile_lab_report(
#         './Evolution of Resistance to Antibiotics in Bacteria Lab Report.pdf',
#         './AppendixA.pdf',
#         './AppendixB.pdf',
#         'compiled_lab_report.pdf'
# )

compile_all_pdfs(['Sensory Lab Report.pdf', 'table181.pdf', 'table184.pdf', 'references.pdf'])
```
